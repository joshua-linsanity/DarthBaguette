Adjust the parameters of the two normal distributions to approximate the desired properties and calculate sample statistics.



```python
import numpy as np
import matplotlib.pyplot as plt

stats = dict()
trackers = ['min', 'max', 'mean', 'median', 'std']
for tracker in trackers: stats[tracker] = 0

bimodal_samples = None
for i in range(100000):
    samples1 = np.random.normal(loc=50, scale=13, size=22)
    samples2 = np.random.normal(loc=80, scale=7, size=22)
    bimodal_samples = np.concatenate([samples1, samples2])

    # Calculate and print sample statistics
    t = 1
    if abs(np.min(bimodal_samples) - 21) > t: continue
    if abs(np.max(bimodal_samples) - 99) > t: continue 
    if abs(np.mean(bimodal_samples) - 64.68) > t: continue
    if abs(np.median(bimodal_samples) - 67.25) > t: continue
    if abs(np.std(bimodal_samples) - 18.08) > t: continue
    
    stats['min'] += np.min(bimodal_samples)
    stats['max'] += np.max(bimodal_samples)
    stats['mean'] += np.mean(bimodal_samples)
    stats['median'] += np.median(bimodal_samples)
    stats['std'] += np.std(bimodal_samples, ddof=1)
    
    plt.hist(bimodal_samples, bins=44)
    break

for key in stats: stats[key]
print(f"min: {stats['min']}")
print(f"max: {stats['max']}")
print(f"mean: {stats['mean']}")
print(f"median: {stats['median']}")
print(f"std: {stats['std']}")

# Calculate percentile
score = 74
print(np.percentile(bimodal_samples, [25,50,60]))
```
